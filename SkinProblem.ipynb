```python
from ultralytics import YOLO
model = YOLO("yolov8s-seg.pt")
```

```python
results = model.train(
    data="/content/drive/MyDrive/SkinProblem/data.yaml",
    epochs=50,
    imgsz=640,
    name="skin-seg",
    project="/content/drive/MyDrive/runs"
)
```

```python
!cp /content/drive/MyDrive/runs/skin-seg8/weights/best.pt /content/drive/MyDrive/SkinProblem/
!cp /content/drive/MyDrive/runs/skin-seg8/weights/last.pt /content/drive/MyDrive/SkinProblem/
```

```python
# Load your saved model
model = YOLO("/content/drive/MyDrive/SkinProblem/best.pt")
```

```python
# Run validation on VALID split
model.val(data="/content/drive/MyDrive/SkinProblem/data.yaml", split="val")
```

```python
# Run validation on TEST split
model.val(data="/content/drive/MyDrive/SkinProblem/data.yaml", split="test")
```

```python
results = model.predict(source="/content/drive/MyDrive/SkinProblem/test/images",
                        save=True,  # saves predictions
                        conf=0.25)  # confidence threshold
```
